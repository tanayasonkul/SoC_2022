```python
def gcd (x,y):
    if x>y:
        small = y
    else:
        small = x
    for i in range(1, small+1):
        if ((x % i == 0) and (y % i == 0)):
            gcd = i
    return gcd
```

```python
def mod_inv (a,b):
    for c in range(1,b):
        if (a*c) % b == 1:
            return c
    return None
```

```python
def coprimes(lambda_n):
    l = []
    for x in range (2,lambda_n):
        if gcd(x, lambda_n) == 1 and mod_inv(x, lambda_n) != None:
            l.append(x)
    for x in l:
        if x == mod_inv(x, lambda_n):
            l.remove(x)
    return l
```

```python
def encrypt_message(m):
    c = (pow(m,e)) % n
    if c == None: print('No modular multiplicative inverse for block ' + str(m) + '.')
    return c

def decrypt_message(c):
    m = (pow(c,d)) % n
    if m == None: print('No modular multiplicative inverse for block ' + str(c) + '.')
    return m
```

```python
def encrypt_string(s):
    return ''.join([chr(encrypt_message(ord(x))) for x in list(s)])
def decrypt_string(s):
    return ''.join([chr(decrypt_message(ord(x))) for x in list(s)])


m = input("Enter a message to encrypt: ")
c = encrypt_string(m)
print("Encrypted Message is: "+ c)
d = decrypt_string(c)
print("Decrypted Message is: "+ d)
```
